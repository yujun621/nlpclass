```python
merge_df = pd.DataFrame()
for i in range(1,31):
    temp = pd.DataFrame()
    if i < 10:
        temp = pd.read_csv(os.path.join(DATASET_PATH,'2019-04-0'+str(i)+'.csv'),header=None,error_bad_lines=False)
        merge_df=pd.concat([merge_df,temp],join='outer',ignore_index=True)
    else:
        temp = pd.read_csv(os.path.join(DATASET_PATH,'2019-04-'+str(i)+'.csv'.format(i)),header=None,error_bad_lines=False)
        merge_df=pd.concat([merge_df,temp],join='outer',ignore_index=True)
        #merge_df.append(temp.copy())
merge_df
```

```python
merge_df.to_csv("./merge.csv")
```
