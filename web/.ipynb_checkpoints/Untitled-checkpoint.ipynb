```python
import requests
from bs4 import BeautifulSoup
```

```python
NAVER_SEARCH_URL = "https://search.naver.com/search.naver"
query = "무역전쟁"
params = {
    'where': 'news',
    'query': query
}
```

```python
resp = requests.get(NAVER_SEARCH_URL, params=params)
resp.status_code
```

```python
soup = BeautifulSoup(resp.content, 'html.parser')
```
