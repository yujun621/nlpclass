```python
BASE_URL = "https://search.naver.com/search.naver"
query = '호국보훈의달'

params={
    'where':'kin',
    'query':query,
    'kin_start':1
}
link_list=[]
for idx in range(5):
    params['kin_start']=params['kin_start']+idx*10
    resp = requests.get(BASE_URL,params)
    soup = BeautifulSoup(resp.content,'html.parser')
    ul=soup.find('div',class_="kinn section _kinBase").find('ul')
    li_list = ul.find_all('li')

    for i in li_list:
        temp_dic={
        'title':'',
        'Q':'',
        'A':''
        }
        #print('title',i.find('dt').text)
        temp_dic['title']=i.find('dt').find('a').text
        #print('Q',i.find_all('dd')[1].text)
        temp_dic['Q']=i.find_all('dd')[1].text
        #print('A',i.find_all('dd')[2].text)
        temp_dic['A']=i.find_all('dd')[2].text[3:]
        link_list.append(temp_dic)
link_list
```
